# Working with scikit-image

```python
import imageio.v3 as iio
import ipympl
import numpy
import skimage
from matplotlib import pyplot

chair = iio.imread(uri="../data/raw/training/chair.jpg")
the_figure, the_axes = pyplot.subplots()
the_axes.imshow(chair)
iio.imwrite(uri="../data/processed/chair.tif", image=chair)
```

## Resize the image

Get the current size, this is the `shape` of the numpy array.
The first two valued are y and x, then the colours.
So the colours should remain as 3 for RGB.

```python
new_shape = (chair.shape[0] // 10, chair.shape[1] // 10, chair.shape[2])
smaller_chair = skimage.transform.resize(image=chair, output_shape=new_shape)
# Convert to integer values between 0 and 255
smaller_chair = skimage.util.img_as_ubyte(smaller_chair)

other_figure, other_axes = pyplot.subplots()
other_axes.imshow(smaller_chair)
```
